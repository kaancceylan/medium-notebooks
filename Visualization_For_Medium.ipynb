# Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
```

```python
tips = pd.read_csv('tips.csv')
```

# Bar Plots

Here's the matplotlib bar chart. We need to set the axis labels manually, which are done for us in Seaborn and Plotly.

```python
fig = plt.hist(x=tips['day'])
plt.xlabel('Day')
plt.ylabel('Count')
```

Here's the Seaborn bar chart, looking a bit tighter.

```python
sns.histplot(data=tips, x='day')
```

If you pass the variable to y, you can get a horizontal graph, which might help you out if the category labels are overlapping on the graph.

```python
sns.histplot(data=tips, y='day')
```

Lastly the Plotly bar chart. You can select a specific area with your cursor and zoom in to take a closer look. Hover over the data to see more details. Double click to zoom-out. You can also download your plot as a PNG file.

```python
fig = px.bar(tips, x='day')
fig.show()
```

# Line Plots

You can just use plt.plot in Matplotlib and it will generate a lineplot for you by default.

```python
plt.plot(tips['tip'])
```

```python
sns.lineplot(data=tips, x='tip', y='total_bill')
```

```python
px.line(tips, x='total_bill')
```

# Scatter Plots

Let's take a look at the relationship between the total bill and tip amount with the scatter plot. Looks like as the total bill amount increases, the tip usually increases too.

```python
plt.scatter(tips['tip'], tips['total_bill'])
```

Let's get a bit fancy with Seaborn and include a hue parameter passing it the days in the dataset. You can see that the highes bills and tips are mostly on Saturdays.

```python
sns.scatterplot(x=tips['tip'], y=tips['total_bill'], hue=tips['day'])
```

We can get the same plot but a lot bigger with plotly so it's easier to read. You can hover on any point to see the bill and tip amounts and zoom in just like before to take a closer look.

```python
px.scatter(x=tips['tip'], y=tips['total_bill'], color=tips['day'])
```

#Box Plots

You get pretty much the same plot with both Matplotlib and Seaborn here.

```python
sns.boxplot(data=tips, x='total_bill')
```

```python
sns.boxplot(tips['total_bill'])
```

Plotly is again a lot more useful especially with boxplots thanks to the interactive plots by default. You can hover and see the median value, Q1 and Q3 values that I've talked about in the blog post. You can also pinpoint the outlier values directly on the plot.

```python
px.box(x=tips['total_bill'])
```

Now that we've taken a look at all the graph types from the blog post, I'm going to show you some cool tricks you can do with Plotly 

# Something Extra

You can add the points parameter and see all the individual datapoints alongside the boxplot so that you can get a better understanding of the variable's distribution. Though this does take the outlier points away, you can set the points variable to 'outliers' or delete it completely to get them back.

```python
px.box(x=tips['total_bill'], points='all')
```

You can pass a categorical variable to the color argument and get different boxplots for each category. Now the cooles thing you can do here is if you click on the categories on the right side of the plot, you can toggle them on and off on the plot.

```python
px.box(x=tips['total_bill'], points='outliers', color=tips['day'])
```
