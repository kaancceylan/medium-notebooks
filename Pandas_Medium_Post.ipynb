##Slicing and Filtering The Data

You can use loc and iloc to slice the part of the dataset that you want to process seperately or do calculations on.

Printing all the rows where the total_bill is greater than 24.

```python
df.loc[df['total_bill'] > 24]
```

Tip amount for groupsize > 6 and total_bill > 24

```python
df.loc[(df['total_bill'] > 24) & (df['size'] > 5), ['tip']]
```

Can do the same thing with iloc but iloc does not accept the filtering we did with loc, iloc takes integers or boolean arrays and the list function inside the iloc returns an array of boolean values.

```python
print(list(df.total_bill > 24))
```

```python
df_filter_size = df.iloc[list(df.total_bill > 24)]
```

```python
print(df_filter_size)
```

We've got the same result that we got with loc before.
Let's say we want to see the group size distribution of these rows. We can use the value_counts() and pass the 'size' to get that and as you can see, 48 out of the 64 rows had a group size of 3 or more.

```python
df_filter_size.value_counts('size')
```

By writing df['a filtering condition'], we can filter the results of the dataframe. The same as we did with iloc, pass a boolean array to the df by writing the filter inside the df[]. Let's filter the dataset by the smokers.

```python
df[df['smoker']=='Yes']
```

Let's see how much missing value we have in our dataset. For that we'll use the isna() (you can also use isnull() ) function along with the sum() function so that the results are easier the read. isna() detects all the missing values in the dataframe and sum() counts them, giving us the final result. If we use only isna(), we get the same dataframe but the values will be masked with False or True values based on if they're null or not.

```python
df.isna()
```

```python
df.isna().sum()
```

Looks like we don't have any missing values in the dataset. Let's say that we had some missing values in the tip column and we figured that they could be filled with the mean tip value, here's how we'd do it.

Filling the null values with the mean of that column is a common practice if the amount of missing data in the column is < 5%. Using the fillna() method to fill the null values, we pass the mean of the column and set the inplace parameter to True (False by default). Inplace will make sure that the mean value is written over the null values and from now on when we call the dataframe, there will be no null values.

```python
df.fillna(df['tip'].mean(), inplace=True)
```
